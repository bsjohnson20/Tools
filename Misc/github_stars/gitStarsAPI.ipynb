```python
# load env
from dotenv import load_dotenv
from bs4 import BeautifulSoup as bs
import os, tabulate
load_dotenv()
```

```python
from github import Github, UnknownObjectException

# Authentication is defined via github.Auth
from github import Auth

# using an access token
auth = Auth.Token(os.getenv("AUTH_TOKEN"))

# First create a Github instance:

# Public Web Github
g = Github(auth=auth)

# load page.html
with open("page.html","r") as f:
    html = f.read()
soup = bs(html, 'html.parser')
# parse the hrefs from the soup
soup = soup.find_all('a', href=True)
links = [x['href'] for x in soup if x['href'].startswith('http') and 'github' in x['href']]




```

```python
# find stars from all repos and use tabulate to print
from tabulate import tabulate
import re
data = []

links = [x for x in links if 'features' not in x]

matcher = re.compile(r'http(s|)://github.com/[a-zA-Z]+/[a-zA-Z]+')

links = [x for x in links if matcher.match(x)]

for link in links:
    try:
        repo = g.get_repo(link.split('github.com/')[1])
        data.append([repo.full_name, repo.stargazers_count, repo.watchers_count])
    except UnknownObjectException as e:
        # print(e)
        continue
    except Exception as e:
        # print(e)
        continue


```

```python
data = data[1::]
data = sorted(data, key=lambda x: x[1], reverse=True)
print(tabulate(data, headers=["Repo", "Stars","Watchers"], tablefmt="pretty"))
```
